```python
import pandas as pd
import numpy as np
from sklearn.preprocessing
import OrdinalEncoder, OneHotEncoder, StandardScaler, MinMaxScaler

# Defina os nomes das 14 colunas conforme .names
colunas = [
    'age', 'workclass', 'fnlwgt', 'education', 'education-num',
    'marital-status', 'occupation', 'relationship', 'race', 'sex',
    'capital-gain', 'capital-loss', 'hours-per-week', 'native-country',
    'income'
]

# Carregar treino
df_train = pd.read_csv('/content/drive/MyDrive/adult/adult.data', header=None, names=colunas, na_values=' ?')

# Carregar teste
df_test = pd.read_csv('/content/drive/MyDrive/adult/adult.test', header=0, names=colunas, skiprows=1, na_values=' ?')
```

```python
print("Amostra dos dados:")
df_train.shape
```

```python
# Mostrar resumo das colunas, tipos e valores nulos
df_train.info()
```

```python
# Contar quantos valores faltam em cada coluna
df_train.isnull().sum()
```

```python
# Preencher valores ausentes nas colunas workclass, occupation e native-country com a moda
for col in ['workclass', 'occupation', 'native-country']:
    # Calcular a moda para o conjunto de treino e preencher os NaN
    mode_train = df_train[col].mode()[0]
    df_train[col].fillna(mode_train, inplace=True)

    # Calcular a moda para o conjunto de teste e preencher os NaN
    mode_test = df_test[col].mode()[0]
    df_test[col].fillna(mode_test, inplace=True)

# Verificar se ainda há valores ausentes após o preenchimento
print("Valores ausentes em df_train após preenchimento:")
display(df_train.isnull().sum())

print("\nValores ausentes em df_test após preenchimento:")
display(df_test.isnull().sum())
```

```python
# Identificar colunas categóricas
categorical_cols = df_train.select_dtypes(include='object').columns

print("Colunas categóricas:")
print(categorical_cols)

# Mostrar valores únicos em algumas colunas categóricas (escolha algumas importantes)
for col in ['workclass', 'education', 'marital-status', 'occupation', 'relationship', 'race', 'sex', 'native-country', 'income']:
    if col in categorical_cols:
        print(f"\nValores únicos em '{col}':")
        display(df_train[col].unique())
```

```python
# Identificar colunas para One-Hot Encoding (excluindo a coluna alvo 'income')
cols_to_onehot = [col for col in categorical_cols if col != 'income']

# Aplicar One-Hot Encoding nos DataFrames de treino e teste
df_train_encoded = pd.get_dummies(df_train, columns=cols_to_onehot, drop_first=True)
df_test_encoded = pd.get_dummies(df_test, columns=cols_to_onehot, drop_first=True)

print("DataFrame de treino após One-Hot Encoding:")
display(df_train_encoded.head())

print("\nDataFrame de teste após One-Hot Encoding:")
display(df_test_encoded.head())
```

```python
from sklearn.preprocessing import StandardScaler

# Identificar colunas numéricas (excluindo a coluna alvo 'income')
# Precisamos fazer isso nos DataFrames codificados, pois as colunas originais foram substituídas
numerical_cols = df_train_encoded.select_dtypes(include=np.number).columns.tolist()

# Remover a coluna 'income' se ela for numérica após a codificação (o que não deve acontecer com One-Hot)
if 'income' in numerical_cols:
    numerical_cols.remove('income')

print("Colunas numéricas para dimensionamento:")
print(numerical_cols)

# Inicializar o StandardScaler
scaler = StandardScaler()

# Aplicar Standard Scaling nas colunas numéricas do conjunto de treino
df_train_scaled = df_train_encoded.copy() # Crie uma cópia para não modificar o dataframe original inplace
df_train_scaled[numerical_cols] = scaler.fit_transform(df_train_scaled[numerical_cols])

# Aplicar Standard Scaling nas colunas numéricas do conjunto de teste (usando o scaler ajustado no treino)
df_test_scaled = df_test_encoded.copy() # Crie uma cópia
df_test_scaled[numerical_cols] = scaler.transform(df_test_scaled[numerical_cols])

print("\nDataFrame de treino após dimensionamento:")
display(df_train_scaled.head())

print("\nDataFrame de teste após dimensionamento:")
display(df_test_scaled.head())
```

```python
# Mapear a coluna 'income' para valores numéricos
# Assumindo que '<=50K' será 0 e '>50K' será 1
income_mapping = {' <=50K': 0, ' >50K': 1, ' <=50K.': 0, ' >50K.': 1}

df_train_scaled['income'] = df_train_scaled['income'].map(income_mapping)
df_test_scaled['income'] = df_test_scaled['income'].map(income_mapping)


# Separar features (X) e alvo (y) para treino
X_train = df_train_scaled.drop('income', axis=1)
y_train = df_train_scaled['income']

# Separar features (X) e alvo (y) para teste
X_test = df_test_scaled.drop('income', axis=1)
y_test = df_test_scaled['income']

print("Shapes dos conjuntos X e y:")
print("X_train shape:", X_train.shape)
print("y_train shape:", y_train.shape)
print("X_test shape:", X_test.shape)
print("y_test shape:", y_test.shape)

print("\nPrimeiras linhas de X_train:")
display(X_train.head())

print("\nPrimeiras linhas de y_train:")
display(y_train.head())
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Inicializar o modelo
model = RandomForestClassifier(random_state=42)

# Treinar o modelo
model.fit(X_train, y_train)

# Fazer previsões no conjunto de teste
# Reindexar as colunas de X_test para corresponder às colunas de X_train
X_test_reindexed = X_test.reindex(columns=X_train.columns, fill_value=0)

y_pred = model.predict(X_test_reindexed)

# Avaliar o modelo (mostraremos a avaliação no próximo passo)
print("Modelo treinado. Previsões feitas no conjunto de teste.")
```

```python
# Avaliar o modelo
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)

print("\nAcurácia do Modelo:", accuracy)
print("\nRelatório de Classificação:")
print(report)
print("\nMatriz de Confusão:")
print(conf_matrix)
```

```python
# Analisar a relação entre raça e renda
# Agrupar por raça do dataframe original e calcular a média da renda do dataframe escalado
# Crie uma cópia do dataframe escalado para adicionar a coluna 'race' de volta
df_train_with_race = df_train_scaled.copy()
df_train_with_race['race'] = df_train['race']

income_by_race = df_train_with_race.groupby('race')['income'].mean().sort_values(ascending=False)

print("Proporção de renda >50K por raça:")
display(income_by_race)
```
