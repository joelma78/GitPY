```python
#Leitura de arquivo
import pandas as pd
import numpy as np

df_vendas_itens=pd.read_excel('/content/drive/MyDrive/Online Retail.xlsx')
show = df_vendas_itens.head()
```

```python
display(df_vendas_itens.dtypes)
```

```python
# Conversão de tipo
df_vendas_itens.CustomerID = df_vendas_itens.CustomerID.astype(float)

display(df_vendas_itens.dtypes)

```

```python
# Um pouquinho de engenharia de atributos
# Apagando os NaNs
df_vendas_itens.dropna(inplace=True)


```

```python
# Criação da data para agrupamento :)
df_vendas_itens['mes_compra'] = df_vendas_itens.InvoiceDate.dt.strftime('%Y-%m')
display(df_vendas_itens.dtypes)
#Convertendo a coluna para dataTime
df_vendas_itens['mes_compra'] = pd.to_datetime(df_vendas_itens['mes_compra'])
display(df_vendas_itens.dtypes)
```

```python
# Calculando o total da venda
df_vendas_itens['total'] = df_vendas_itens['Quantity'] * df_vendas_itens['UnitPrice']
display(show)
```

```python
# Conte as linhas onde 'total' é menor que zero - negativos
contagem_negativos = (df_vendas_itens['total'] < 0).sum()
print(f"Número de linhas com 'total' negativo: {contagem_negativos}")
```

```python
# Removendo totais negativos
df_vendas_itens = df_vendas_itens[df_vendas_itens['total'] > 0]
```

```python
contagem_negativos = (df_vendas_itens['total'] < 0).sum()
print(f"Número de linhas com 'total' negativo: {contagem_negativos}")

#Exibindo depois de retiradas
```

```python
# Reorganizando as colunas
df_vendas_itens.columns = ['venda_id', 'produto_id', 'descricao', 'quantidade', 'data_venda',
                           'preco', 'cliente_id', 'cliente_pais', 'mes_venda', 'total']

df_vendas_itens = df_vendas_itens[['venda_id', 'mes_venda', 'data_venda', 'cliente_id',
                                   'cliente_pais', 'produto_id', 'descricao', 'quantidade',
                                   'preco', 'total']]

print(df_vendas_itens.dtypes)
df_vendas_itens.shape
```
